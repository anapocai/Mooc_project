```python
# Garantir que a coluna 'date' est√° no formato datetime
df_merged['date'] = pd.to_datetime(df_merged['date'], errors='coerce')

# Extrair o ano e o m√™s da coluna 'date'
df_merged['year'] = df_merged['date'].dt.year
df_merged['month'] = df_merged['date'].dt.month

# Agrupar os dados por pa√≠s, ano e m√™s, somando as mortes e outros dados de interesse
df_monthly = df_merged.groupby(['country', 'year', 'month']).agg({
    'total_deaths': 'sum',
    'new_deaths': 'sum',
    'total_cases': 'sum',
    'new_cases': 'sum',
    'gdp_per_capita': 'mean',  # Para PIB per capita, pega a m√©dia
    'human_development_index': 'mean',  # Para HDI, pega a m√©dia
    'extreme_poverty': 'mean',  # M√©dia de pobreza extrema
    'diabetes_prevalence': 'mean',  # M√©dia de preval√™ncia de diabetes
    'hospital_beds_per_thousand': 'mean',  # M√©dia de leitos hospitalares
    'population': 'last'  # Pega o valor da popula√ß√£o no √∫ltimo dia do m√™s
}).reset_index()

# Verificar os primeiros registros para confirmar que a transforma√ß√£o foi bem-sucedida
print(df_monthly.head())
```

```python
# Fazendo o merge entre df_monthly e df_group_df para adicionar a coluna 'Group'
df_merged_group = pd.merge(df_monthly, group_df, how='left', on='country')

# Agrupar os dados por grupo e somar as colunas de interesse
df_grouped = df_merged_group.groupby(['Group', 'year', 'month']).agg({
    'total_deaths': 'sum',
    'new_deaths': 'sum',
    'total_cases': 'sum',
    'new_cases': 'sum',
    'population': 'sum',  # Soma da popula√ß√£o (pode ser ajustada para m√©dia, caso necess√°rio)
    'gdp_per_capita': 'mean',  # M√©dia de PIB per capita
    'human_development_index': 'mean',  # M√©dia de HDI
    'extreme_poverty': 'mean',  # M√©dia de pobreza extrema
    'diabetes_prevalence': 'mean',  # M√©dia de preval√™ncia de diabetes
    'hospital_beds_per_thousand': 'mean',  # M√©dia de leitos hospitalares
}).reset_index()

# Verificar os primeiros registros para confirmar que a transforma√ß√£o foi bem-sucedida
print(df_grouped.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group the data by Group, Year, and Month, summing up the total cases
df_grouped_cases = df_grouped.groupby(['Group', 'year', 'month'])['total_cases'].sum().reset_index()

# Create a 'date' column for proper time representation
df_grouped_cases['date'] = pd.to_datetime(df_grouped_cases[['year', 'month']].assign(day=1))

# Plot the line chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_grouped_cases, x='date', y='total_cases', hue='Group', marker='o')

# Titles and labels
plt.title('Total COVID Cases Trend by Group Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Cases', fontsize=12)
plt.legend(title='Country Group')
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python


# Group the data by Group, Year, and Month, summing up the total cases and population
df_grouped_cases = df_grouped.groupby(['Group', 'year', 'month']).agg({
    'total_cases': 'sum',
    'population': 'sum'  # Sum up the population for each group
}).reset_index()

# Calculate the percentage of cases in relation to the population
df_grouped_cases['cases_percentage'] = (df_grouped_cases['total_cases'] / df_grouped_cases['population']) 

# Create a 'date' column for proper time representation
df_grouped_cases['date'] = pd.to_datetime(df_grouped_cases[['year', 'month']].assign(day=1))

# Plot the line chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_grouped_cases, x='date', y='cases_percentage', hue='Group', marker='o')

# Titles and labels
plt.title('Percentage of COVID Cases Relative to Population by Group Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Percentage of Population Infected (%)', fontsize=12)
plt.legend(title='Country Group')
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python

# Group the data by Group, Year, and Month, summing up deaths and cases
df_grouped_mortality = df_grouped.groupby(['Group', 'year', 'month']).agg({
    'total_deaths': 'sum',
    'total_cases': 'sum'
}).reset_index()

# Calculate the mortality rate (avoid division by zero)
df_grouped_mortality['mortality_rate'] = (df_grouped_mortality['total_deaths'] / df_grouped_mortality['total_cases']) * 100
df_grouped_mortality['mortality_rate'].fillna(0, inplace=True)  # Replace NaN values with 0

# Create a 'date' column for proper time representation
df_grouped_mortality['date'] = pd.to_datetime(df_grouped_mortality[['year', 'month']].assign(day=1))

# Identify the first row for the European group in 2020
mask = (df_grouped_mortality['Group'] == 'European Union (EU) and Regional Powers') & (df_grouped_mortality['year'] == 2020)

# Drop the first occurrence of this row
df_grouped_mortality = df_grouped_mortality[~(mask & (df_grouped_mortality.index == mask.idxmax()))]

# Plot the line chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_grouped_mortality, x='date', y='mortality_rate', hue='Group', marker='o')

# Titles and labels
plt.title('COVID-19 Mortality Rate by Group Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Mortality Rate (%)', fontsize=12)
plt.legend(title='Country Group')
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Realizar o merge entre os dois DataFrames usando a coluna 'Country'
merged_turism = pd.merge(df_turism, group_df, how='inner', left_on='Country', right_on='country')

# Visualizar as primeiras linhas do DataFrame resultante
print(merged_turism.head())
```

```python
merged_turism = merged_turism.dropna()
merged_turism = merged_turism.drop_duplicates()
```

```python
# Converter a coluna 'DATE' para datetime
merged_turism['DATE'] = pd.to_datetime(merged_turism['DATE'])

# Extrair ano e m√™s
merged_turism['Year'] = merged_turism['DATE'].dt.year
merged_turism['Month'] =merged_turism['DATE'].dt.month

# Agrupar por ano e m√™s e somar a quantidade
monthly_data = merged_turism.groupby(['Group','Year', 'Month'])['QUANTITY'].sum().reset_index()

# Visualizar o resultado
print(monthly_data)
```

```python
import matplotlib.pyplot as plt

# Group by Group and Year, then sum the quantities
grouped_evolution = monthly_data.groupby(['Group', 'Year'])['QUANTITY'].sum().reset_index()

# Plot the evolution of groups over the years
plt.figure(figsize=(10, 6))

for group in grouped_evolution['Group'].unique():
    group_data = grouped_evolution[grouped_evolution['Group'] == group]
    plt.plot(group_data['Year'], group_data['QUANTITY'], label=group)

plt.xlabel('Year')
plt.ylabel('Quantity of Tourists')
plt.title('Evolution of Groups Over the Years')
plt.legend(title='Groups')
plt.grid(True)
plt.show()
```

```python
import plotly.express as px

# Filter data for Europe and group by continent and country
europe_data = merged_turism[merged_turism['CONTINENT'] == 'EU']

# Loop through each year and create a treemap
for year in europe_data['Year'].unique():
    year_data = europe_data[europe_data['Year'] == year]
    fig = px.treemap(year_data, 
                     path=['CONTINENT', 'country'], 
                     values='QUANTITY', 
                     title=f"Tourism in Europe by Country in {year}")
    fig.show()


```

```python
# Filter data for the last 6 months of 2019 and up to 2021
covid_period = merged_turism[(merged_turism['DATE'] >= '2019-06-01') & (merged_turism['DATE'] <= '2021-12-31')]

# Extract year and month
covid_period['Year'] = covid_period['DATE'].dt.year
covid_period['Month'] = covid_period['DATE'].dt.month

# Group by year, month, and continent, and sum the quantity
covid_grouped = covid_period.groupby(['CONTINENT', 'Year', 'Month'])['QUANTITY'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Plot the data for each continent
for continent in covid_grouped['CONTINENT'].unique():
    continent_data = covid_grouped[covid_grouped['CONTINENT'] == continent]
    plt.plot(continent_data['Year'].astype(str) + '-' + continent_data['Month'].astype(str), continent_data['QUANTITY'], label=continent)

plt.xlabel('Year-Month')
plt.ylabel('Quantity of Tourists')
plt.title('Tourism Trends by Continent Before and After COVID-19')
plt.xticks(rotation=45)
plt.legend(title='Continent')
plt.grid(True)
plt.show()
```

```python
import pycountry

# Fun√ß√£o para mapear nome do pa√≠s para c√≥digo ISO-3
def get_iso3(country_name):
    try:
        return pycountry.countries.lookup(country_name).alpha_3
    except:
        return None  # se n√£o encontrar, deixa vazio pra gente filtrar depois

# Aplicar no DataFrame
europe_yearly['ISO3'] = europe_yearly['Country'].apply(get_iso3)
europe_yearly = europe_yearly.dropna(subset=['ISO3'])

```
